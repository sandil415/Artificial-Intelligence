```python
def sumList(numList):
  print(f"The sum of the list is: {sum(numList)}")

def averageList(numList):
  print(f"The average of the list is: {sum(numList) / len(numList)}")

def maxList(numList):
  print(f"The max of the list is: {max(numList)}")

def minList(numList):
  print(f"The min of the list is: {min(numList)}")

def main():
    try:
        print("Choose an operation:")
        print("1. Sum")
        print("2. Average")
        print("3. Maximum")
        print("4. Minimum")

        choice = input("Enter your choice (1-4): ")

        numbers_input = input("Enter numbers separated by spaces: ")

        # Convert input string to list of floats
        numbers = [float(num) for num in numbers_input.split()]

        if not numbers:
            raise ValueError("The list cannot be empty.")

        if choice == "1":
          sumList(numbers)

        elif choice == "2":
          averageList(numbers)

        elif choice == "3":
          maxList(numbers)
        elif choice == "4":
          minList(numbers)

        else:
            print("Invalid operation selected.")

    except ValueError:
        print("Error: Please enter valid numbers separated by spaces.")


# Run the program
main()

```

```python
def access_nested_element(lst, indices):
    for i in indices:
        lst = lst[i]
    return lst

access_nested_element([1, [2, [3, 4], 5], 6], [1, 1, 0])
```

```python
def sum_nested(lst):
  sum = 0
  for i in lst:
    if isinstance(i, list):
      sum += sum_nested(i)
    else:
      sum += i
  return sum

print(sum_nested([[1, 2], [3, [4, 5]], 6],))
```

```python
def remove_ele(lst, n):
  result = []

  for i in lst:
    if i == n:
      continue

    if isinstance(i, list):
      result.append(remove_ele(i, n))
    else:
      result.append(i)

  return result


print(remove_ele([[1, 2], [3, [4, 5]], 6], 6))
```

```python
def find_max(lst):
    maxNum = float('-inf')  # works with negative numbers

    for i in lst:
        if isinstance(i, list):
            returnedVal = find_max(i)
            if returnedVal > maxNum:
                maxNum = returnedVal
        else:
            if i > maxNum:
                maxNum = i

    return maxNum

print(find_max([[1, 2], [3, [4, 5]], 6]))
```

```python
def count_occurences(lst, elem):
  sum = 0
  for i in lst:
    if isinstance(i, list):
      sum += count_occurences(i, elem)
    else:
      if i == elem:
        sum += 1
  return sum

count_occurences([[1, 2], [2, 3], [2, 4]], 2)
```

```python
def deep_flatten(lst):
  newList = []
  for i in lst:
    if isinstance(i, list):
      newList.extend(deep_flatten(i))
    else:
      newList.append(i)
  return newList

deep_flatten([[1, 2], [3, [4, 5]], 6])
```

```python
def average_nested(lst):
  newList = []
  for i in lst:
    if isinstance(i, list):
      newList.extend(deep_flatten(i))
    else:
      newList.append(i)
    return newList

average_nested([[1, 2], [3, [4, 5]], 6])
```
